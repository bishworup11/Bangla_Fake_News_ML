**Creating a function to convert the text in lowercase, remove the extra space, special characters., url and links**

```python
def wordopt(article):
    article	 = article.lower()
    article	 = re.sub('\[.*?\]', '', article)
    article = re.sub("\\W"," ",article)
    article = re.sub('https?://\S+|www\.\S+', '', article)
    article = re.sub('<.*?>+', '', article)
    article = re.sub('[%s]' % re.escape(string.punctuation), '', article)
    article = re.sub('\n', '', article)
    article = re.sub('\w*\d\w*', '', article)
    return article
```

```python
df["article"] = df["article"].apply(wordopt)
```

**Defining dependent and independent variable as x and y**

```python
x = df["article"]
y = df["label"]
```

**Splitting the dataset into training set and testing set.**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

**Convert text to vectors**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

### **Model Testing With Manual Entry**
### **News**

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"article":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["article"] = new_def_test["article"].apply(wordopt)
    new_x_test = new_def_test["article"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GBC = GBC.predict(new_xv_test)
    pred_RFC = RFC.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_lable(pred_LR[0]),
                                                                                                              output_lable(pred_DT[0]),
                                                                                                              output_lable(pred_GBC[0]),
                                                                                                              output_lable(pred_RFC[0])))
```

```python
news = str(input())
manual_testing(news)
```
